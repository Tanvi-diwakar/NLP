```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

#Loading pre-trained multilingual model

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

#creating sentimental analysis pipeline

```python
sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
def analyze_sentiment(texts):
    for text in texts:
        result = sentiment_analyzer(text)
        print(f"Text: {text}\nSentiment: {result[0]['label']}, Score: {result[0]['score']:.4f}\n")
```

```python
texts = [
"I hate this model.",
"Je déteste ce système.",
"Este modelo es terrible.",
"Dieses Modell ist schlecht.",
"このモデルは嫌いです"
]
```

```python
analyze_sentiment(texts)
```
